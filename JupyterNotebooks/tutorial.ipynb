```python
# Import python packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib auto
```

```python
# Open image
filename = "/home/l1nx/NIWE_SkyImages/20190108/20190108_090000.jpg"
filename = "/media/l1nx/A54F-A5EE/rbr_test/20190109_064950.jpg"
# Use OpenCV to read the image (OpenCV reads an image as BGR (Blue, Green, Red))
img = cv2.imread(filename)
print(img)
```

```python
# Convert image

# Convert from BGR -> RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Create Grayscale Image (1-dimensional image with pixel intensity)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# floating point image
tmp = np.array(img, dtype=np.float32)

# RBR image (divide by non-0 number)
rbr = img[:,:,0] / (img[:,:,2]+0.0001)
```

```python
# Visualize image with matplotlib
plt.imshow(img_gray)
plt.show()

# Visualize with given color range and colorbar
plt.imshow(rbr, vmin=0.8, vmax=1.2)
plt.colorbar()
plt.show()
```

```python
# Plot histogram
plt.hist(rbr.flatten(), range=[0,1.3], bins=100)
plt.show()
```
